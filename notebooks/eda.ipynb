# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Cargamos CSV con datos combinados

```python
data_file = 'data/combined_data.csv'
df = pd.read_csv(data_file)
```

## Eliminar columnas no útiles

```python
columns_to_drop = ['_id_x', '_id_y', 'start_time', 'end_time', 'conf', 'hits', 'age', 'time_since_update', 'state', 'time']
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

## Mostrar las primeras filas del DataFrame

```python
print("Primeras filas del DataFrame:")
df.head()
```

## Mostrar información del DataFrame

```python
print("\nInformación del DataFrame:")
df.info()
```

## Estadísticas descriptivas

```python
print("\nEstadísticas descriptivas:")
df.describe()
```

## Visualización de distribuciones

```python
plt.figure(figsize=(12, 8))
sns.histplot(df['cord_x'], kde=True)
plt.title('Distribución de cord_x')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(df['cord_y'], kde=True)
plt.title('Distribución de cord_y')
plt.show()
```

## Visualización de correlaciones

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de calor de correlaciones')
plt.show()
```

## Visualización de relaciones entre variables

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='cord_x', y='cord_y', data=df, hue='track_id', palette='viridis')
plt.title('Relación entre cord_x y cord_y')
plt.show()
```

## Visualización de variables expandidas

```python
mean_columns = [f'mean_{i}' for i in range(8)]
cov_columns = [f'cov_{i}' for i in range(8)]
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(df[mean_columns])
plt.title('Relación entre columnas mean')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(df[cov_columns])
plt.title('Relación entre columnas covariance')
plt.show()
```

## Guardamos el nuevo CSV

```python
output_file = 'data/eda_data.csv'
df.to_csv(output_file, index=False)
```
